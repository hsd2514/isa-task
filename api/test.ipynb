```python
#lets build a function to accept data and make it machine learning ready
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#here is our refrence data
df=pd.read_csv("../data/Uncleaned_employees_final_dataset (1).csv")
df.head()
 
```

```python
#now lets build a function to accept data and make it machine learning ready
def data_prep(data):
    #lets drop the employee id column
    data=data.drop(['employee_id'],axis=1)

    #lets use pandas api to convert the data into categorical data
    for column in data.columns:
        if column is not pd.api.types.is_numeric_dtype:
            data[column]=data[column].astype('category')
            data[column]=data[column].cat.codes

    return data
    
```

```python
data_prep(df)
```
